```python
import numpy as np
import matplotlib.pyplot as plt
import numpy_financial as npf
```

```python
periods = np.arange(1,30,1)
```

```python
rate = .05
amount = 100
```

```python
fv = npf.fv(rate,periods,0,-amount,0)
print(fv)
```

```python
plt.plot(periods,fv)
plt.show()
```

```python
pv = npf.pv(rate,periods,0,-amount,0)
```

```python
plt.plot(periods,pv)
plt.show()
```

```python
fig = plt.figure()
```

```python
ax1 = fig.add_subplot(211)
ax1.plot(periods,fv)
ax1.set_xlabel("Periods")
ax1.set_ylabel("Future Value")
ax1.set_title("Compounding")
```

```python
ax2 = fig.add_subplot(212)
ax2.plot(periods,pv)
ax2.set_xlabel("Periods")
ax2.set_ylabel("Future Value")
ax2.set_title("Compounding")
```

```python
fig.tight_layout()
```
